```python
%matplotlib notebook
import matplotlib.pyplot as plt
import islr
import util
import sigpy as sp
import sigpy.mri.rf as rf
```

# Set parameters

```python
n = 64  # Pulse length
tb = 8  # Time bandwidth
d1 = 0.01  # Passband ripple
d2 = 0.01  # Stopband ripple
```

# Design RF pulse using the original SLR

```python
b1_slr = rf.dzrf(n=n, tb=tb, ptype='ex', ftype='pm', d1=d1, d2=d2, cancel_alpha_phs=True)
```

```python
figs = util.plot_rf(b1_slr, balanced=True)
```

# Design RF pulse using the improved SLR

```python
b1_islr = islr.dzrf(n=n, tb=tb, ptype='ex', d1=d1, d2=d2)
```

```python
figs = util.plot_rf(b1_islr, balanced=True)
```
